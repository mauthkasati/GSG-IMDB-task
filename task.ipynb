```python
import pandas as pd
```

import pandas library

```python
data_frame = pd.read_csv("C:\datasets\imdb_top_1000.csv")
data_frame
```

load data set to data_frame from attached directory

```python
chosen_columns = ['Genre', 'IMDB_Rating', 'Director', 'No_of_Votes']
df2 = data_frame[chosen_columns]
df2
```

choose needed columns to achieve the tasks, and drop the others. 
I save it in a new data frame df2

```python
df2 = df2.dropna(axis=0)
df2
```

remove rows containing missing values

```python
genre_counts = df2['Genre'].value_counts()
genre_counts
```

finding counts of each genre

```python
most_popular_genre = genre_counts.index[0]
most_popular_genre
```

most popular genre is the first row, since the genre_counts is sorted desc

```python
genre_counts_df = pd.DataFrame(genre_counts).reset_index()
genre_counts_df.columns = ['Genre', 'Count']
genre_counts_df
```

creating a new dataframe containing two columns Genre, Count

```python
import matplotlib.pyplot as plt
```

import matplotlib.pyplot to plot the needed plots

```python
genre_counts_df.plot(x='Genre', y='Count', kind='bar')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre')
plt.show()
```

plotting all genres with related countes

```python
genre_counts_df2 =genre_counts_df.head(20)
genre_counts_df2.plot(x='Genre', y='Count', kind='bar')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre')
plt.show()
```

plotting top 20 genres with related counts

```python
director_counts = df2['Director'].value_counts()
director_counts
```

finding counts of each director

```python
most_popular_director = director_counts.index[0]
most_popular_director
```

most popular director is the first row, since the director_counts is sorted desc

```python
director_counts_df = pd.DataFrame(director_counts).reset_index()
director_counts_df.columns = ['Director', 'Count']
director_counts_df
```

```python
director_counts_df.plot(x='Director', y='Count', kind='bar')
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Director with Most Top-rated Movies/TV Shows')
plt.show()
```
